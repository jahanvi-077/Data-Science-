## loop in python
* Looping generally means repeating a sequence of actions again and again until a certain condition is met.





| Loop Type         | Description                                            | Example Use Case                                                         |
| ----------------- | ------------------------------------------------------ | ------------------------------------------------------------------------ |
| **For Loop**      | Repeats a block of code a specific number of times.    | Iterating through a list or repeating an action a fixed number of times. |
| **While Loop**    | Repeats as long as a condition remains true.           | Running code until the user gives a certain input.                       |
| **Do-While Loop** | Runs code at least once before checking the condition. | Menu systems or input validation.                                        |
| **For-Each Loop** | Iterates through elements in a collection.             | Reading items in an array or list.                                       |



## For loop 
* A for loop in Python is used to repeat a block of code for each item in a sequence (like a list, range, string, or tuple).

## While 
* A while loop in Python is used to repeat a block of code as long as a condition remains true.

```python
count = 1

while (count <= 5):
    print(count)
    count += 1
```

```python
lst = [1,2,3]
for x in lst:
    print(x)
   
    if x<=2:
        print(x)
        
    
```
