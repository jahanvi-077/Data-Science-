### Single Inheritance

```python
class Collage:
    pass


class Student(Collage):
    pass
```

```python
class Collage:
    account_number = 846685
    collage_name = "jecrc"
class Student(Collage):
    pass

st=Student()
```

```python
class Collage:
    account_number = 52846
    collage_name = "jecrc"
    def account(self):
       print(f"my acccount number is {self.account_number}")
class Student(Collage):
    student_name = "janvi"

st = Student()
```

```python
st.account()
```

```python
st.student_name
```
