#### Condiional Statements
A conditional statement in Python is a way to make your program choose what to do based on whether a condition is true or false.
#### Types of conditional statements
| Type                           | Description                                                       | Python Example                                                                     |
| ------------------------------ | ----------------------------------------------------------------- | ---------------------------------------------------------------------------------- |
| **if**                         | Executes code **only if** the condition is true.                  | `python if x > 10: print("Big") `                                                  |
| **if–else**                    | Runs one block if the condition is true, otherwise another block. | `python if x > 10: print("Big") else: print("Small") `                             |
| **elif (else-if)**             | Used to check **multiple conditions** one after another.          | `python if score >= 90: print("A") elif score >= 80: print("B") else: print("C") ` |
| **Nested if**                  | An **if inside another if** for more specific checks.             | `python if logged_in: if is_admin: print("Access granted") `                       |
| **Shorthand (single-line) if** | A compact version of a simple if statement.                       | `python print("Yes") if x > 5 else print("No") `                                   |

#### looping in python 
Looping in Python means repeating a block of code again and again until a certain condition is met or until you finish going through items in a sequence (like a list).

#### types of loops in python
| Loop Type             | What It Does                                                               | Easy Example                                                |
| ---------------------- | --------------------------------------------------------------------------- | ----------------------------------------------------------- |
| **for loop**           | Repeats code for **each item** in a list, range, string, etc.               | `python for i in range(5): print(i) `                       |
| **while loop**         | Repeats code **as long as** a condition remains true.                       | `python while x < 5: print(x); x += 1 `                     |
| **nested loop**        | A loop **inside another loop**, used for repeated patterns.                 | `python for i in range(3): for j in range(2): print(i, j) ` |
| **loop with break**    | Stops the loop **early** when a condition is met.                           | `python for n in nums: if n < 0: break `                    |
| **loop with continue** | Skips the current step and goes to the **next loop cycle**.                 | `python for n in nums: if n == 0: continue `                |
| **loop with else**     | The `else` block runs **after the loop finishes normally** (without break). | `python for i in range(3): print(i) else: print("Done") `   |

#### Inheritance in python
* Inheritance in Python allows one class to take (inherit) the properties and methods of another class.
* It helps you reuse code, reduce repetition, and create a relationship between classes.
#### Types of inheritance
| Type of Inheritance          | Description                                                                                                | Simple Example                                                                |
| ---------------------------- | ---------------------------------------------------------------------------------------------------------- | ----------------------------------------------------------------------------- |
| **Single Inheritance**       | One child class inherits from **one** parent class.                                                        | `python class A: pass class B(A): pass `                                      |
| **Multiple Inheritance**     | A child class inherits from **more than one** parent class.                                                | `python class A: pass class B: pass class C(A, B): pass `                     |
| **Multilevel Inheritance**   | A class inherits from a parent, and another class inherits from that child. (Grandparent → Parent → Child) | `python class A: pass class B(A): pass class C(B): pass `                     |
| **Hierarchical Inheritance** | Multiple child classes inherit from the **same** parent class.                                             | `python class A: pass class B(A): pass class C(A): pass `                     |
| **Hybrid Inheritance**       | A combination of two or more inheritance types.                                                            | `python class A: pass class B(A): pass class C(A): pass class D(B, C): pass ` |


```python
#single inheritance
class Collage:
    pass
#here its one class

class Student(Collage):
    pass
#and its another
```

```python
#multilevel inheritance
class Grandfather:
    account_number = 3284895
class Father(Grandfather):
    phone_number = 84688789 
class Child(Father):
    pin_number = 35784655

chi = Child()
```

```python
#multiple inheritance
class Grandfather:
    grandfather_name = "govind"
class Uncle(Grandfather):
    uncle_name = "Ramesh"    
class Tau(Grandfather):
    Tau_name = "satyanarayan" 
tau_obj =Tau()
tau_obj.grandfather_name
```
